```python
#create a function that removes digit
def remove_digit(x):
    if x == np.nan:
        return
    return ''.join([i for i in x if not i.isdigit()]).strip()
```

```python
full_merged_data["Released_month"] = full_merged_data[["Released"]].apply(lambda x : remove_digit(str(x)))
```
