```python
df = df_250_train[["Title", "Actors"]].copy()
df["actors_list"] = df["Actors"].apply(lambda x : x.split(", "))
df
```

```python
df_explode = df.explode(column = "genre_list")
df_explode
```

```python
df_explode["genre_proc"] = df_explode["genre_list"].str.lower().str.split(" ").apply(lambda x: " ".join(sorted(x)))
```

```python
df_explode
```

```python
actor_dict = {}
for actors in df_explode["genre_proc"]:
    actor_dict[actors] = actor_dict.get(actors, 0) +1
actor_dict
```

```python
len(actor_dict.keys())
```

```python
df_explode["freq"] = df_explode["genre_proc"].apply(lambda x: actor_dict[x])
```

```python
df_explode
```

```python
df_freq_genre = df_explode[df_explode["freq"] <=2]
```

```python
df_dummy = pd.get_dummies(df_freq_genre, columns = ["genre_proc"])
df_dummy
```
