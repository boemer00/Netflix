# Regex for Genre

```python
genre = full_data_merged["Genre"].fillna("")

# vectorizer = CountVectorizer(max_features=2)

# X = vectorizer.fit_transform(genre)

# print(vectorizer.get_feature_names())
```

```python
vectorizer = CountVectorizer(token_pattern="[a-zA-Z][a-z -]+", max_features=5)

X = vectorizer.fit_transform(genre)

print(vectorizer.get_feature_names())
```

```python
print(X.toarray())
print(X.toarray().shape)
```

```python
pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
```

```python
full_data_merged.join(pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names()))
```

```python
full_data_merged = full_data_merged.fillna("")
full_data_merged[full_data_merged["Director"].str.contains("co-director")]["Director"]
```

# Regex for actors 

```python
actores = full_data_merged["Actors"].fillna("")

vectorizer = CountVectorizer(token_pattern="[a-zA-Z][a-z -]+", max_features=10)

X = vectorizer.fit_transform(actors)

print(vectorizer.get_feature_names())

# print(X.toarray())

```

```python
pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
```

```python
full_data_merged.join(pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names()))
```
