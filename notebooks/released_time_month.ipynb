```python
df["Released"] = df["Released"].fillna("Non Available")
```

```python
#create a function that removes digit
def remove_digit(x):
    return ''.join([i for i in x if not i.isdigit()]).strip()
```

```python
df["Released_month"] = df["Released"].apply(remove_digit)
```

```python
import re
def clean_runtime(x):
            counter = 0
            if 'h' in x:
                counter += int(x[0]) * 60
                x = re.sub('.*h', '', x).strip()
            x = x.replace('min', '').replace(',','').strip()
            counter += int(x)
            return counter
```

```python
df["Runtime"] =  df[["Runtime"]].apply(clean_runtime)
```
