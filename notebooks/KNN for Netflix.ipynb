# Create a function that returns the 5 most similar songs

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
```

```python
url = "/Users/davidguenoun/Downloads/ML_spotify_data.csv"
df = pd.read_csv(url)
df
```

```python
def most_similar_songs(song):
    X = df.drop(columns = ["name", "artists"])
    y = df["tempo"]
    knn_model = KNeighborsRegressor().fit(X,y)
    song = df.loc[df['name'] == song]
    ind_list = list(knn_model.kneighbors(song.drop(columns=['name','artists']),n_neighbors=6)[1][0])
    list_song = df.iloc[ind_list, :]["name"].tolist()
    return list_song[1:]

most_similar_songs("Working Without A Net")
```
