```python
import pandas as pd
import requests
import imageio
import matplotlib.pyplot as plt 
from IPython import display
```

# First Method

```python
def movie_poster(Title):
    
    #define base URL 
    url_OMDB = "http://www.omdbapi.com/"
    url_Poster = "http://img.omdbapi.com/"
    
    params_OMDB = {"apikey" : "ef5507df", "t" : Title}
    
    #return movie describition
    response = requests.get(url_OMDB,params = params_OMDB).json()
    
    #extract from the movie describtion the "ImdbID key"
    imdbID = response["imdbID"]
    
    #assign ImdbID key to our params_Poster
    params_Poster = {"apikey" : "ef5507df", "i" : imdbID}
    
    #call the Poster API with the ImdbID
    response = requests.get(url_Poster,params = params_Poster)
    
    #converting bytes object into images
    image = imageio.imread(response.content)
    image = plt.imshow(image)
    
    return image
    
```

```python
movie_poster("Inception")
```

# Second Method

```python
def movie_Poster_2(Title):

    url_OMDB = "http://www.omdbapi.com/"

    params_OMDB = {"apikey" : "ef5507df", "t" : Title}

    response = requests.get(url_OMDB,params = params_OMDB).json()

    Poster = response["Poster"]
    
    Poster = display.Image(Poster)
    
    return Poster
```

```python
movie_Poster_2("Inception")
```
